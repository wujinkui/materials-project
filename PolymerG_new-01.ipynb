```python
import pandas as pd
database = pd.read_csv('E:/科研/data/fomular_3.csv',engine= 'python')
repeatunit = database.RE

def geturl(unitpath):
    urlset = []
    units = []
    database = pd.read_csv(unitpath, engine='python')
    repeatunit = database.Structure
    Host = 'https://www.polymergenome.org/predict/index.php?m=11&kt='
    for unit in repeatunit:
        url = Host + unit
        urlset.append(url)
        units.append(unit)
    return urlset,units
```
